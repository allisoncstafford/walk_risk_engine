```python
import geopandas as gpd
from shapely.geometry import Point
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
boroughs = gpd.read_file('boroughboundaries.shp')
print (boroughs)
```

```python
from shapely import geometry
p1 = Point(-74.06, 40.659)
p2 = Point(-74.270, 40.659)
p3 = Point(-74.27, 40.914)
p4 = Point(-74.06, 40.914)
pointlist = [p1, p2, p3, p4]
poly = gpd.GeoSeries(geometry.Polygon([[p.x, p.y] for p in pointlist]))

p5 = Point(-74.053, 40.688) #40.688310, -74.053245
p6 = Point(-74.06, 40.688)
p7 = Point(-74.06, 40.919)
p8 = Point(-73.913, 40.919) #40.919607, -73.953894
pointlist2 = [p5, p6, p7, p8]
poly2 = gpd.GeoSeries(geometry.Polygon([[p.x, p.y] for p in pointlist2]))
```

```python
base = boroughs.plot(color='white', edgecolor='black')
poly.plot(ax=base, color='blue')
poly2.plot(ax=base, color='red')
```
